````python
import pandas as pd
import glob
import os

base_path = "/content/drive/MyDrive/Internship projects/Phone Pe project - Labmentix Internship/data-csv"
output_path = "/content/drive/MyDrive/Internship projects/Phone Pe project - Labmentix Internship/cleaned_data"

os.makedirs(output_path, exist_ok=True)

def extract_metadata(file_path):
    parts = file_path.lower().split("/")

    year = None
    quarter = None
    state = None
    country = None

    for part in parts:
        if part.isdigit() and len(part) == 4:
            year = int(part)
        elif part.endswith(".csv"):
            quarter = int(part.replace(".csv", ""))

    if "state" in parts:
        idx = parts.index("state")
        if idx + 1 < len(parts):
            state = parts[idx + 1]

    if "country" in parts:
        idx = parts.index("country")
        if idx + 1 < len(parts):
            country = parts[idx + 1]

    return year, quarter, state, country


def process_folder(main_folder, sub_folder):
    output_file = f"{output_path}/{main_folder}_{sub_folder}.csv"

    # ✅ Skip if already exists
    if os.path.exists(output_file):
        print(f"⏭️ Skipping (already exists): {output_file}")
        return

    pattern = f"{base_path}/{main_folder}/{sub_folder}/**/*.csv"
    files = glob.glob(pattern, recursive=True)

    dfs = []

    for file in files:
        try:
            df = pd.read_csv(file)

            year, quarter, state, country = extract_metadata(file)

            df["year"] = year
            df["quarter"] = quarter
            df["state"] = state
            df["country"] = country

            dfs.append(df)

        except Exception as e:
            print(f"Error in {file}: {e}")

    if dfs:
        final_df = pd.concat(dfs, ignore_index=True)
        final_df.to_csv(output_file, index=False)

        print(f"✅ Created: {output_file} | Shape: {final_df.shape}")


# 🔥 Run for all 9 combinations
main_folders = ["aggregated", "map", "top"]
sub_folders = ["insurance", "transaction", "user"]

for main in main_folders:
    for sub in sub_folders:
        process_folder(main, sub)
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Handling missing values
df["state"] = df["state"].fillna("All")
df["country"] = df["country"].fillna("india")
````

````python
# Standardize Categorical Values and handle missing values across all dataframes
import os
import pandas as pd

for file_name in files:
    file_path = os.path.join(output_path, file_name)
    df = pd.read_csv(file_path)

    print(f"\nProcessing {file_name} for standardization...")

    # Define a list of columns that might exist and need standardization
    columns_to_standardize_lower = ["type", "name", "brand", "state", "category", "source", "level", "country"]
    columns_to_standardize_title = ["state"] # only state needs title case (after lowercasing)

    # Apply lowercase standardization
    for col in columns_to_standardize_lower:
        if col in df.columns and df[col].dtype == 'object':
            df[col] = df[col].str.lower()

    # Apply title case standardization for 'state'
    for col in columns_to_standardize_title:
        if col in df.columns and df[col].dtype == 'object':
            # Ensure 'state' is converted to title case after lowercasing
            df[col] = df[col].str.title()

    # Handle missing values for state and country, which were identified as having NaNs
    if "state" in df.columns:
        df["state"] = df["state"].fillna("All") # 'All' is Title case now for consistency
    if "country" in df.columns:
        df["country"] = df["country"].fillna("India") # 'India' is Title case now for consistency

    # Save the modified dataframe back to its original path
    df.to_csv(file_path, index=False)
    print(f"Standardization and missing value handling applied and saved for {file_name}")
````

````python
# Correcting data types across all dataframes
import os
import pandas as pd

for file_name in files:
    file_path = os.path.join(output_path, file_name)
    df = pd.read_csv(file_path)

    print(f"\nProcessing {file_name} for data type correction...")

    # Convert 'year' and 'quarter' to integer if they exist
    if "year" in df.columns:
        df["year"] = df["year"].astype(int)
    if "quarter" in df.columns:
        df["quarter"] = df["quarter"].astype(int)

    # Convert 'count' and 'amount' to numeric, coercing errors
    if "count" in df.columns:
        df["count"] = pd.to_numeric(df["count"], errors='coerce')
    if "amount" in df.columns:
        df["amount"] = pd.to_numeric(df["amount"], errors='coerce')

    # Save the modified dataframe back to its original path
    df.to_csv(file_path, index=False)
    print(f"Data type correction applied and saved for {file_name}")
````

````python
# Feature engineering
for file_name in files:
    df = pd.read_csv(f"{output_path}/{file_name}")

    print(f"\nProcessing {file_name}...")
    # Creating new features
    if "year" in df.columns and "quarter" in df.columns:
        df["date"] = df["year"].astype(str) + "-Q" + df["quarter"].astype(str)

    if "amount" in df.columns and "count" in df.columns:
        df["avg_transaction_value"] = df["amount"] / df["count"].replace(0, pd.NA)

    # Determining growth
    if "count" in df.columns:
        df = df.sort_values(by=["state", "year", "quarter"])
        df["growth"] = df.groupby("state")["count"].pct_change()

    # Handling outliers
    if "amount" in df.columns:
        upper_limit = df["amount"].quantile(0.99)
        df["amount"] = df["amount"].clip(upper=upper_limit)

    # Save back
    df.to_csv(f"{output_path}/{file_name}", index=False)
````

#### Chart - 10

````python
# Chart - 10 visualization code
import pandas as pd
import plotly.graph_objects as go

# Growth calculation
df["growth"] = df.groupby("state")["count"].pct_change()
growth_year = df.groupby("year")["growth"].mean().reset_index()

# Trend assignment (fixed logic)
growth_year["trend"] = 0

for i in range(len(growth_year)):
    if i == 0:
        growth_year.loc[i, "trend"] = 1
    else:
        if growth_year.loc[i, "growth"] >= growth_year.loc[i-1, "growth"]:
            growth_year.loc[i, "trend"] = 1
        else:
            growth_year.loc[i, "trend"] = -1


fig = go.Figure()

# Build segments
start = 0

for i in range(1, len(growth_year)):
    if growth_year.loc[i, "trend"] != growth_year.loc[i-1, "trend"]:

        segment = growth_year.iloc[start:i+1]

        color = "green" if growth_year.loc[i-1, "trend"] == 1 else "red"

        fig.add_trace(go.Scatter(
            x=segment["year"],
            y=segment["growth"],
            mode="lines+markers",
            line=dict(color=color),
            showlegend=False
        ))

        start = i

# last segment
segment = growth_year.iloc[start:]
color = "green" if growth_year.loc[len(growth_year)-1, "trend"] == 1 else "red"

fig.add_trace(go.Scatter(
    x=segment["year"],
    y=segment["growth"],
    mode="lines+markers",
    line=dict(color=color),
    showlegend=False
))

fig.update_layout(
    title="Overall Growth Trend (Year-wise Comparison)",
    xaxis_title="Year",
    yaxis_title="Growth"
)

fig.show()
````

##### 1. Why did you pick the specific chart?

A line chart is used to represent the overall growth trend as it effectively captures changes over time. It helps in visualizing fluctuations in growth rates across years, making it easier to identify patterns such as periods of increase, decline, or stability in transaction activity.
